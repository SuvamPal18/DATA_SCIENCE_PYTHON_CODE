# Visualization the Car Dataset

### Import Librarys

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load the dataset

```python
data = pd.read_csv("C:\\Users\\suvam\\Desktop\\CELEBAL TECHNOLGOY INTERNSHIP\\car data.csv")
print(data.head())
```

```python
data.shape
```

```python
data.describe
```

### Check for missing values

```python
print(data.isnull().sum())

for column in data.select_dtypes(include=['float64', 'int64']).columns:
    data[column].fillna(data[column].median(), inplace=True)

for column in data.select_dtypes(include=['object']).columns:
    data[column].fillna(data[column].mode()[0], inplace=True)
```

### Visualization the Dataset

```python
data["Selling_Price"].hist()
```

```python
data["Present_Price"].hist()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data["Selling_Price"], data["Present_Price"], color='skyblue')
```

```python
plt.title('Total Sales per Product')
plt.xlabel("Present_Price")
plt.ylabel("Selling_Price")
plt.show
```

```python
data.plot(kind="scatter", x="Selling_Price", y="Present_Price", alpha=0.1)
```

```python
from pandas.plotting import scatter_matrix
attributes = ["Selling_Price", "Present_Price", "Kms_Driven"]
scatter_matrix(data[attributes], figsize=(12, 8))
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```
